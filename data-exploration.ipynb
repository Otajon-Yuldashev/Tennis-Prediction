```python
import pandas as pd

years = [2020, 2021, 2022, 2023, 2024, 2025]
dfs = []

for year in years:
    url = f"http://tennis-data.co.uk/{year}/{year}.xlsx"
    print(f"Loading {year}...")
    df_year = pd.read_excel(url)
    dfs.append(df_year)

df = pd.concat(dfs, ignore_index=True)
print(f"\nTotal matches: {len(df)}")
df.head()
```

```python
print("Info about data: ")
print(df.shape)
print("details regarding data: ")
df.info()
print("how many null values per column: ")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
print(df['Surface'].value_counts())
print()
print(df['Court'].value_counts())
print()
print(df['Series'].value_counts())
print()
print(df['Round'].value_counts())
print()
print(df['Best of'].value_counts())
```

```python
print("Winner Rank stats:")
print(df['WRank'].describe())
print()
print("Loser Rank stats:")
print(df['LRank'].describe())
```

```python
higher_ranked_won = (df['WRank'] < df['LRank']).sum()
total = len(df)
percentage = higher_ranked_won / total

print(f"Higher-ranked player won: {higher_ranked_won}/{total}")
print(f"Baseline accuracy: {percentage:.2%}")
```

```python
print(f"Earliest match: {df['Date'].min()}")
print(f"Latest match: {df['Date'].max()}")
print(f"Duplicate rows: {df.duplicated().sum()}")
print(df['Comment'].value_counts())
```
